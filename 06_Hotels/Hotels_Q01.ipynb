```python
import numpy as np
import pandas as pd
```

```python
#Loading the dataset
hotel_file = 'DataFiles/city_hotel_features.txt'
features_file = 'DataFiles/features.txt'
hotel = pd.read_csv(hotel_file, delimiter='\t')
hotel.head()
```

```python
#COnverting the features column into dummy variables
hotel_dummies = hotel.Features.str.get_dummies(sep=' ')
hotel = pd.concat([hotel,hotel_dummies],axis=1)
hotel.drop('Features',axis=1,inplace=True)
hotel.head()
```

```python
hotel['Total'] = hotel.sum(axis=1)
hotel.head()
```

```python
#features = ['005','024','025']
features = ['076']
ones = list(np.ones(len(features),dtype=int))
ones
```

```python
def find_hotel_with_features(hotel,features):
    bool_filter = []
    for values in hotel[features].values:
        bool_filter.append(list(values) == ones)
    return hotel[bool_filter].sort_values('Total',ascending=False).head(10)
```

```python
find_hotel_with_features(hotel,features)
```
