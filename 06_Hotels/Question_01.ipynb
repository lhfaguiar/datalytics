```python
# Necassary imports
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from operator import itemgetter
```

```python
# readfile
file_path = '~/code/Datalytics/DataFiles/Hotels/FiltredDataFile.csv'
df = pd.read_csv(file_path, header=0)
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
# file data info
df.info()
```

```python
# describe data
df.describe()
```

```python
def get_hotels_from_features(input_feature_list):
        
    matched_hotel = []
    
    for index, row in df.iterrows():
        found_row = [f for f in input_feature_list if row[f] != 0]
        if len(input_feature_list) == len(found_row):
             # matched_hotel.append([index, row['Total'], row['Hotel_Name'], row['City_Name']])
            matched_hotel.append([index])
            
    if len(matched_hotel) > 10:
        matched_hotel = sorted(matched_hotel, key=itemgetter(1), reverse=True)
        matched_hotel = matched_hotel[:9]
        
    top_matched_hotels = [i[0] for i in matched_hotel]
     # top_matched_hotels = matched_hotel
     # top_matched_hotels.insert(0, ['ID','Total Features', 'Hotel', 'City'])
    
    return top_matched_hotels
```

```python
get_hotels_from_features(['005', '024', '025'])
```

### Problem 2: Find top 5 important features for given city.

```python
def get_city_top_features(df, city):
    d1 = df.groupby('City_Name').sum().loc[city].sort_values(ascending=False)[1:]
    return pd.DataFrame({'Feature':d1.index, 'Count':d1.values}).head(5)
```

```python
get_city_top_features(df, 'washington_dc')
```
