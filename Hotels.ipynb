```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
hotel_file = '~/code/Datalytics/DataFiles/Hotels/city_hotel_features.txt'

hotel = pd.read_csv(hotel_file, delimiter='\t')
hotel.head()
hotel['feature_list'] = [f.split() for f in hotel['Features']]
hotel['feature_list'] = hotel['feature_list']
```

```python
features_file = '~/code/Datalytics/DataFiles/Hotels/features.txt'
features = pd.read_csv(features_file, delimiter='\t')
features.head()
# features_length = len(features['ID']) - 1
```

```python
s = Series(hotel['feature_list'])
features_list = pd.get_dummies(s.apply(pd.Series).stack()).sum(level=0)
```

```python
a = pd.concat([hotel, features_list], axis=1)
a.head()
```

```python
a.drop(['Features', 'feature_list'], axis=1, inplace=True)
```

```python
a.head()
```

```python
import copy

b = copy.deepcopy(a)
b['Total'] = a.sum(axis=1)
```
